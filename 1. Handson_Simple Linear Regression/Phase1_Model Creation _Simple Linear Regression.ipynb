```python
#pandas is a library that is used of database related work
import pandas as pd 
```

```python
#read the input csv file and store in a variable name
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#print the variable
dataset
```

```python
#Split of ip vs op data
# save i/p data in a variable
independent=dataset[["YearsExperience"]]
```

```python
# save o/p data in a variable
dependent=dataset[["Salary"]]
```

```python
# print i/p data
independent
```

```python
# print o/p data
dependent
```

```python
# use sklearn library to spilt the train data v test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# create the model
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_test,y_test)
```

```python
#create new variables to see the weight and bias
weight= regressor.coef_
bias=regressor.intercept_
```

```python
#print weight
weight
```

```python
#print bias
bias
```

```python
# for test data find the predicted values using the model we created
y_pred=regressor.predict(x_test)
```

```python
# find the r2_score to compare the actual vs expected results
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
# print the score to see if is close to 1 to find if the model is working efficiently
r_score
```
